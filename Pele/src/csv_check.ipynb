```python
import pandas as pd
import glob
import os
```

```python
folder_path = '../data'
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
if not csv_files:
    print(f"'{folder_path}' No CSV file in forlder.")
else:
    print(f"{len(csv_files)}'s CSV file founded'.\n")

    for file_path in csv_files:
        file_name = os.path.basename(file_path)
        print(f"========================================")
        print(f"File name: {file_name}")
        print(f"----------------------------------------")
        
        try:
            df = pd.read_csv(file_path, nrows=5)
            
            print(f"{'Header (Header)':<25} | {'Data Type (Dtype)':<15}")
            print("-" * 45)
            for column, dtype in df.dtypes.items():
                print(f"{str(column):<25} | {str(dtype):<15}")
            
        except Exception as e:
            print(f"Error occurred while reading the file: {e}")
            
        print(f"========================================\n")
```
